```python
#Create a new column of Maritial Statu by using maritial_divorced, maritial_single, maritial_Together, maritial_Widow, maritial_married
def get_marital_status(row):
    if row['marital_Divorced'] == 1:
        return 'Divorced'
    elif row['marital_Married'] == 1:
        return 'Married'
    elif row['marital_Single'] == 1:
        return 'Single'
    elif row['marital_Together'] == 1:
        return 'Together'
    elif row['marital_Widow'] == 1:
        return 'Widow'
    else:
        return 'Unknown'
data['Marital'] = data.apply(get_marital_status, axis=1)
```

```python
#Distribution of Mnttotal by Maritial status 
sns.boxplot(
    x = "Marital",
    y = "MntTotal",
    showmeans=True, 
    data = data
)
plt.xlabel('Marital Status')
plt.ylabel('Mnttotal')
plt.title('Distribution of Mnttotal by Marital Status', bbox=dict(facecolor='lightgray'))
plt.xticks(rotation=45)
plt.show()
```

We observe that married and together women spent more than single 

```python
#Create a new column of Education by using educatioin_phd, education_master, education_basic, education_graduation
def education(row):
    if row['education_Basic'] == 1:
        return 'Basic'
    elif row['education_Graduation'] == 1:
        return 'Graduation'
    elif row['education_PhD'] == 1:
        return 'PHD'
    elif row['education_Master'] == 1:
        return 'Master'
    else:
        return 'Unknown'
data['Education'] = data.apply(education, axis=1)
```

```python
#Distribution of Mnttotal by Education Qualification
sns.boxplot(
    x = "Education",
    y = "MntTotal",
    showmeans=True,
    data = data
)
plt.xlabel('Education Qualification')
plt.ylabel('Mnttotal')
plt.title('Distribution of Mnttotal by Education Qualification', bbox=dict(facecolor='lightgray'))
plt.xticks(rotation=45)
plt.show()
```

boxplot shows that masters spent more than other 

```python
#Distribution of Income by Education Qualification 
sns.boxplot(
    x = "Education",
    y = "Income",
    showmeans=True,
    data = data
)
plt.xlabel('Education Qualification')
plt.ylabel('Income')
plt.title('Distribution of Income by Education Qualification', bbox=dict(facecolor='lightgray'))
plt.xticks(rotation=45)
plt.show()
```

Boxplot shows that phd people have more income in comparison to other

Customer segmentation

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[['Income', 'MntTotal']]) 
```

```python
kmeans = KMeans(n_clusters=4) #we use 4 cluster for k means custering.. 
kmeans.fit(scaled_data)
data['Segment'] = kmeans.labels_
```

```python
plt.scatter(data['Income'], data['MntWines'], c=data['Segment'])
plt.xlabel('Income')
plt.ylabel('Monthly Wine Purchase')
plt.title('Customer Segments by Income and Wine Purchase')
plt.show()
```

```python
plt.scatter(data['Income'], data['MntFruits'], c=data['Segment'])
plt.xlabel('Income')
plt.ylabel('Monthly Fruit Purchase')
plt.title('Customer Segments by Income and Fruit Purchase')
plt.show()
```

```python
plt.scatter(data['Income'], data['MntFishProducts'], c=data['Segment'])
plt.xlabel('Income')
plt.ylabel('Monthly Fish Purchase')
plt.title('Customer Segments by Income and Fish Purchase')
plt.show()
```

```python
plt.scatter(data['Income'], data['MntMeatProducts'], c=data['Segment'])
plt.xlabel('Income')
plt.ylabel('Monthly Meat Purchase')
plt.title('Customer Segments by Income and Meat Purchase')
plt.show()
```

```python
plt.scatter(data['Income'], data['MntGoldProds'], c=data['Segment'])
plt.xlabel('Income')
plt.ylabel('Monthly Gold Purchase')
plt.title('Customer Segments by Income and Gold Purchase')
plt.show()
```

```python
plt.scatter(data['Income'], data['MntSweetProducts'], c=data['Segment'])
plt.xlabel('Income')
plt.ylabel('Monthly Sweet Purchase')
plt.title('Customer Segments by Income and Sweet Purchase')
plt.show()
```

```python
cluster_sizes = data.groupby('Segment')[['MntTotal']].count().reset_index()
plt.figure(figsize=(8,6))
sns.barplot(x='Segment', y='MntTotal', data=cluster_sizes, palette = 'viridis')
plt.title('Cluster sizes')
plt.xlabel('Segment')
plt.ylabel('MntTotal')
```

In cluster 2 , high mnttotal
